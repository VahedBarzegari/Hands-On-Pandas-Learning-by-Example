üü¶ 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

üü¶ 2. Sample Dataset

```python
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 90, 95],
    'Science': [88, 92, 84],
    'English': [91, 85, 89]
})

df
```

üü¶ 3. Basic melt() ‚Äî Convert Wide ‚Üí Long

```python
melted = pd.melt(df, id_vars='Student')
melted
```

üü¶ 4. Rename Melted Columns (var_name & value_name)

```python
melted_named = pd.melt(
    df,
    id_vars='Student',
    var_name='Subject',
    value_name='Score'
)

melted_named
```

üü¶ 5. Melt Multiple ID Columns

```python
#add a new column
df2 = df.copy()
df2['Class'] = ['A', 'A', 'B']

#melt with two columns
pd.melt(df2, id_vars=['Student', 'Class'],
        var_name='Subject', value_name='Score')
```

üü¶ 6. Melt Multiple Value Columns

```python
df3 = pd.DataFrame({
    'Student': ['Ali', 'Reza'],
    'Math_Score': [80, 90],
    'Math_Grade': ['B', 'A'],
    'Science_Score': [85, 92],
    'Science_Grade': ['B+', 'A-']
})

df3
```

```python
melted_multi = pd.melt(
    df3,
    id_vars='Student',
    value_vars=['Math_Score', 'Math_Grade', 'Science_Score'],
    var_name='Subject_Metric',
    value_name='Value'
)

melted_multi
```

üü¶ 7. Extract Information After Melt

```python
split_df = melted_multi.copy()
split_df[['Subject', 'Metric']] = split_df['Subject_Metric'].str.split('_', expand=True)
split_df
```

üü¶ 8. Keep Only Selected Columns (Dropping Others)

```python
pd.melt(df, value_vars=['Math', 'Science'])
```
