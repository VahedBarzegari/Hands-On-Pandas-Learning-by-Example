ðŸŸ¦ 1. Import & Sample Data

```python
import pandas as pd

data = {
    "trip_id": [1, 2, 3, 4, 5],
    "timestamp": [
        "2024-01-15 08:30:45",
        "2024-03-20 14:10:05",
        "2024-06-01 05:15:00",
        "2024-09-10 22:50:30",
        "2024-12-25 12:00:00"
    ]
}

df = pd.DataFrame(data)

df["timestamp"] = pd.to_datetime(df["timestamp"])
df
```

ðŸŸ¦ 2. Extract Year, Month, Day

```python
df["year"] = df["timestamp"].dt.year
df["month"] = df["timestamp"].dt.month
df["day"] = df["timestamp"].dt.day

df
```

ðŸŸ¦ 3. Extract Hour, Minute, Second

```python
df["hour"] = df["timestamp"].dt.hour
df["minute"] = df["timestamp"].dt.minute
df["second"] = df["timestamp"].dt.second

df
```

ðŸŸ¦ 4. Day of Week & Day Name

```python
df["weekday_num"] = df["timestamp"].dt.weekday  # Monday=0
df["weekday_name"] = df["timestamp"].dt.day_name()

df
```

ðŸŸ¦ 5. Extract Week Number

```python
df["week_number"] = df["timestamp"].dt.isocalendar().week
df
```

ðŸŸ¦ 6. Extract Quarter

```python
df["quarter"] = df["timestamp"].dt.quarter
df
```

ðŸŸ¦ 7. Extract Month Name

```python
df["month_name"] = df["timestamp"].dt.month_name()
df
```

ðŸŸ¦ 8. Extract Day of Year

```python
df["day_of_year"] = df["timestamp"].dt.dayofyear
df
```

ðŸŸ¦ 9. Extract Is Weekend

```python
df["is_weekend"] = df["timestamp"].dt.weekday >= 5
df
```

ðŸŸ¦ 10. Feature Engineering Example (Real World)

```python
df["time_period"] = pd.cut(
    df["hour"],
    bins=[0, 6, 10, 16, 20, 24],
    labels=["Night", "Morning Peak", "Midday", "Evening Peak", "Late Night"],
    right=False
)

df
```
