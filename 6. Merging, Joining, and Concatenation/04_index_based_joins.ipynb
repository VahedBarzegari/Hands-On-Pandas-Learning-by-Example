ðŸŸ¦ 1. Import Libraries

```python
import pandas as pd
```

ðŸŸ¦ 2. Sample DataFrames

```python
df_left = pd.DataFrame({
    "value_left": [10, 20, 30]
}, index=["A", "B", "C"])

df_right = pd.DataFrame({
    "value_right": [100, 200, 300]
}, index=["A", "B", "D"])
```

ðŸŸ¦ 3. Basic Index Join (Default = Left Join)

```python
df_left.join(df_right)
```

ðŸŸ¦ 4. Inner Join on Index

```python
df_left.join(df_right, how="inner")
```

ðŸŸ¦ 5. Outer Join (Union of All Indexes)

```python
df_left.join(df_right, how="outer")
```

ðŸŸ¦ 6. Joining Multiple DataFrames at Once

```python
df_extra = pd.DataFrame({
    "value_extra": [1, 2, 3]
}, index=["A", "B", "C"])

df_left.join([df_right, df_extra], how="outer")
```

ðŸŸ¦ 7. Joining on Columns by Setting Index First

```python
dfA = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Ali", "Sara", "Omid"]
}).set_index("id")

dfB = pd.DataFrame({
    "id": [1, 3, 4],
    "salary": [5000, 7000, 6000]
}).set_index("id")

dfA.join(dfB, how="left")
```
