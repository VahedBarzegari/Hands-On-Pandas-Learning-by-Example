ðŸŸ¦ 1. Import Required Library

```python
from pandas import DataFrame
import pandas as pd
```

ðŸŸ¦ 2. Example Dataset

```python
df = DataFrame({
    "route_id": [10, 10, 20, 20, 30, 30],
    "direction": [0, 0, 0, 1, 0, 1],
    "delay_min": [3, 5, 7, 2, 1, 4],
    "trip_duration": [25, 30, 45, 50, 20, 35]
})

df
```

ðŸŸ¦ 2. Multiple Aggregations on a Single Column

```python
df.groupby("route_id")["delay_min"].agg(["mean", "sum", "max"])
```

ðŸŸ¦ 3. Aggregating Different Columns Differently

```python
df.groupby("route_id").agg({
    "delay_min": ["mean", "max"],
    "trip_duration": "sum"
}).reset_index()
```

ðŸŸ¦ 4. Renaming Output Columns for Clarity

```python
df.groupby("route_id").agg(
    avg_delay=("delay_min", "mean"),
    max_delay=("delay_min", "max"),
    total_duration=("trip_duration", "sum")
).reset_index()
```
