ðŸŸ¦ 1. Import Required Library

```python
from pandas import DataFrame
import pandas as pd
```

ðŸŸ¦ 2. Example Dataset

```python
df = DataFrame({
    "route_id": [10, 10, 20, 20, 30, 30],
    "direction": [0, 0, 0, 1, 0, 1],
    "day_type": ["Weekday", "Weekend", "Weekday", "Weekend", "Weekday", "Weekend"],
    "delay_min": [3, 5, 7, 2, 1, 4],
    "trip_duration": [25, 30, 45, 50, 20, 35]
})

df
```

ðŸŸ¦ 2. Grouping by Multiple Columns

```python
df.groupby(["route_id", "direction"])["delay_min"].mean()
```

ðŸŸ¦ 3. Aggregating Multiple Columns on Multi-Group

```python
df.groupby(["route_id", "direction"]).agg(
    avg_delay=("delay_min", "mean"),
    total_duration=("trip_duration", "sum")
)
```

ðŸŸ¦ 4. Resetting Index to Flatten

```python
df.groupby(["route_id", "direction"]).agg(
    avg_delay=("delay_min", "mean"),
    total_duration=("trip_duration", "sum")
).reset_index()
```

ðŸŸ¦ 5. Comparing Nested Groups

```python
df.groupby(["route_id", "day_type"])["delay_min"].mean().reset_index()
```
